```python
import re

fileName = '/Users/sbk/Documents/Jobs/Professional_Info.txt'

with open(fileName, 'r', encoding='latin-1') as f:
    lines = [line.strip() for line in f]
    cv_data = {}

    for line in lines:
        # Match the field name and its content
        match = re.match(r"#(.*?):\s*(.*)", line)
        if match:
            field_name = match.group(1).strip()
            content = match.group(2).strip()

            # Split content into subfields by ';'
            subfields = content.split(';')
            subfield_data = {}

            for subfield in subfields:
                subfield = subfield.strip()
                if ':' in subfield:
                    # Split subfield into key-value pairs by ':'
                    key, value = map(str.strip, subfield.split(':', 1))
                    subfield_data[key] = value
                elif subfield:
                    # Handle subfields without key-value pairs
                    subfield_data[subfield] = None

            cv_data[field_name] = subfield_data

    # Display the structured CV data
    for field, subfields in cv_data.items():
        print(f"{field}:")
        for subfield, value in subfields.items():
            print(f"  {subfield}: {value}")
```

```python
# Random job announcement
job_announcement = """
Job Title: Chef in a French Restaurant
Location: Stockholm, Sweden
Job Description:
We are looking for an experienced chef to join our team at a prestigious French restaurant in Stockholm.
The ideal candidate should have a passion for French cuisine, excellent culinary skills, and the ability to
work in a fast-paced environment. Responsibilities include preparing high-quality dishes, managing kitchen staff,
and ensuring food safety standards are met. The candidate should also have experience in menu planning and
inventory management.
Qualifications:
- Minimum 5 years of experience as a chef in a fine dining restaurant
- Strong knowledge of French cuisine and cooking techniques
- Ability to lead and motivate a team
- Excellent communication and organizational skills
- Culinary degree or equivalent certification preferred
- Fluent in Swedish and English
- Availability to work evenings and weekends
"""

# Generate the tailored prompt
prompt = f"""
Based on the following job announcement:

{job_announcement}

And my professional information:

{cv_data}

Design a tailored resume, highlighting my relevant skills, experiences, and certifications. 
Do not include what it doesn't match the profile of the job. Be humble but confident in my abilities. 
Provide actionable tips to make my application stand out for this role.
"""

# Call the AI function
tip = get_ai_tip(prompt)

# Display the AI-generated tip
display(Markdown(f"### AI Tip:\n\n{tip}"))
```
