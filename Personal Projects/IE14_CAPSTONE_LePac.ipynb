**-Creacion de dataset:** Se crea una lista agregando numeros aleatoreos de la lista **l** hasta llenar con los datos necesarios, luego se calcula el promedio de la lista creada y si no cumple estar en el intervalo deseado, se vuelve a crear nuevamente hasta que se logre.
La lista **l** se crea con las opciones discretas del 1 al 7, pero dando mayor opciones a ciertos elementos, para crear la distribución deseada.

### Creando nuestro test de hipótesis

Desarrollar un test de hipótesis de 1 muestra para al menos 1 de los 3 datasets. Aquí Ud. puede formular la hipótesis a probar. Por ejemplo, puede ser contra un target definido por la industria o retail.



### <font color='red'>__Respuesta__:<br> 

La prueba de hipótesis:
* __Hipótesis nula__ (__$H_0$__): la proproción de las personas que han manifestado satisfacción por el producto/servicio (*Top 2 Box Score*) es **igual** a la satifaccion actual (línea base), es decir, un 29%.
* __Hipótesis alternativa__ (__$H_1$)__: la proproción de las personas que han manifestado satisfacción por el producto/servicio (*Top 2 Box Score*) es **mayor** a la satifaccion actual (línea base), es decir, un 29%.

Planteamiento estadístico:
$$H_0: \quad p = 0.29$$
$$H_1: \quad p > 0.29$$

### __2. Determinar el tamaño de la muestra__

* Nivel de significancia / $\alpha$ = 0.05
* $\beta$ = 0.2
* Potencia de prueba ($1- \beta$) = 0.8 
* Tamaño del efecto deseado
* Varianza de la población ($\sigma$)

Para determinar el tamaño del efecto, aquí es necesario hablar con el cuerpo experto en esta materia o con los directores del negocio donde estemos operando. Para este ejemplo vamos a suponer que **el target que define la empresa** es un cambio de versión de producto de al menos **10 puntos porcentuales brutos**, de tal forma que con la nueva versión el porcentaje de satifacción producto/servicio sea de al menos un **39%** (0.39).

* Efecto mínimo a determinar: 0.10

```python
p0 = 0.29 # línea base de proporción rebote
efecto_bruto = 0.1 # valor minimo (diferencia) a detectar como mejora del estudio
p1 = p0 + efecto_bruto
alpha = 0.05
beta = 0.20

N = ((p0*(1-p0)*( stats.norm.ppf(1-alpha) + stats.norm.ppf(1-beta)* np.sqrt( (p1*(1-p1)) /(p0*(1-p0))))**2) /
     (p1 - p0)**2)

print(f'Tamaño mínimo muestra: {N:5.1f}')
```

- Punto N° 1, ya que las muestras de Le Pac y Capel tienen varianza desiguales.

```python
# Análisis post hoc utilizando t Student, con varianza desigual

## 1) cargar datos muestras: tamaño (n), media (xbarra) y sigma (S)
# muestra 1 (control)
n_1 = N_Capel
xbarra_1 = mu_Capel
S_1 = des_Capel

# muestra 2 (experimental)
n_2 = N_Lepac
xbarra_2 = mu_Lepac
S_2 = des_Lepac

## 2) Determinar Nivel de Significancia
alpha = 0.05

## 3) Estimar el IC de la diferencia de medias

# Estimar grados de libertad cuando varianzas poblacionales no son iguales
gl = (S_1**2 /n_1 + S_2**2 /n_2)**2  /  (((S_1**2/n_1)**2 / (n_1-1)) + ((S_2**2/n_2)**2 / (n_2-1)))
print(f'Estimación gl: {gl:4.2f}')

# calcular T critico 2 colas con Scipy
Tcritico = stats.t.ppf(1 - (alpha/2), df=gl)
print (f'T critico: {Tcritico:5.3f}')


# Estimar ponderado de la varianza poblacional (S pooled) =>  #NO APLICA para calcular IC ya que Varianzas son diferentes
# Se aplica la varianza combinada para calcular 'd Cohen' en el último paso, por esta razón se calcula aquí:
var = ( S_1**2 *(n_1 - 1) + S_2**2 *(n_2 - 1)) / (n_1 + n_2 - 2)
S = np.sqrt(var)
print(f'Varianza combinada: {var:5.3f}')
print(f'Sigma población: {S:5.3f}')


# como las varianzas son diferentes, el cálculo del 'Sigma diferencias  medias' es de la siguiente forma:
S_dif = np.sqrt( (S_1**2 / n_1) + (S_2**2 / n_2) )
print(f'S diferencia medias: {S_dif:5.3f}')

# IC diferencia medias al 1-alpha:
LI = (xbarra_2 - xbarra_1) - Tcritico * S_dif
LS = (xbarra_2 - xbarra_1) + Tcritico * S_dif
print(f'IC diferencia medias al {1-alpha:5.3f} NC: {LI:5.3f}    {LS:5.3f}')

# otra forma con Scipy
IC = stats.t.interval(1-alpha, loc=xbarra_2 - xbarra_1, scale=S_dif, df=gl)
print(f'IC diferencia medias al {1-alpha:5.3f} NC: {IC[0]:5.3f}   {IC[1]:5.3f}')
```

```python
# Otra forma de estimar grados de libertad con varianzas desiguales es con la la ecuación Welch–Satterthwaite. 
# Nótese que la verdadera distribución de este estadístico de hecho depende (ligeramente) de dos varianzas desconocidas.
gll = (S_1**2 /n_1 + S_2**2 /n_2)**2  /  (((S_1**2/n_1)**2 / (n_1-1)) + ((S_2**2/n_2)**2 / (n_2-1))) - 2
print(f'Grados de Libertad: {gll:5.3f}')
```

```python
## 4) plantear hipótesis
 # H0:  Mu 1 = Mu 2
 # H1:  Mu 1 ≠ Mu 2

# regla decisión (2 colas)
 # no rechazar H0 si: -Tcritico <= T <= Tcritico
 # rechazar HO si: T < -Tcritico o T > Tcritico

## 5) Calcular el estadístico T crítico a emplear
Tcritico = Tcritico #fue obtenido en paso anterior

## 6) Determinar el tipo de test, una o dos colas.
# como la hipòtesis alternativa es del tipo “diferente de” se trata de un test de dos colas
# Nota: Se recomienda utilizar el contraste bilateral para estos análisis dado que éstos 
# requieren una mayor diferencia para detectar una misma potencia

## 7 ) Calcular el estadístico T de prueba
# como varianzas no son iguales, aqui el denominador es el sigma de diferencias calculado en paso #3
T = ( xbarra_2 - xbarra_1 - 0) / S_dif
print(f'T prueba: {T:5.3f}')

## 8) Calcular el p-value y comparación con el nivel de significancia establecido.
# calcular p-value
pvalue = 1-(stats.t.cdf(T, df=(n_1 + n_2 -2))) # T prueba se encuentra en cola izquierda
print(f'p-value (significancia estadística): {pvalue*2}') # 2 colas multiplicar por 2

## 9) Concluir prueba de hipótesis:
# como T prueba > Tcritico o como p-value menor a alpha, se rechaza H0, las medias son diferentes.
```

```python
## 10) Calcular la potencia de la prueba (1 - beta)

# buscar el valor en días correspondiente al T critico
x = 0 + (Tcritico * S_dif)# por la derecha
# print('X promedio para T=',Tcritico,': ',x)

# otra forma mas simple con Scipy usando ppf
x_b = stats.t.ppf((1-alpha/2), df=n_1 + n_2 - 2 , loc=0, scale=S_dif)
# print('X promedio para T=',Tcritico,': ',x_b)


# Calcular Beta (error tipo II)
# forma simple con Scipy
beta = stats.t.cdf(x, df=n_1 + n_2 -2, loc=xbarra_2-xbarra_1, scale=S_dif)
print(f'Beta: {beta:5.3f}')

# otra forma simple para calcular Beta con Scipy
beta1 = stats.t.cdf(Tcritico-T, df= n_1 + n_2 - 2)
print(f'Otra forma beta: {beta1:5.3f}')

# Calcular Potencia estadística (1- beta)
print(f'Potencia de la prueba: {1-beta:5.3f}')


## 11) Calcular el tamaño del efecto
# aquí utilizaremos la desv. combinada (S pooled)
# d de Cohen para T-TEST INDEPENDIENTES: (diferencia medias grupos / S medias 2 sigmas)

dif_bruta = (xbarra_2 - xbarra_1)
dCohen = (xbarra_2 - xbarra_1) / S
print(f'd Cohen: {dCohen:5.3f}')
print(f'Diferencia bruta: {dif_bruta:5.3f}')
```

Las diferencias son estadísticamente significativas (**p-value infimo**), el cambio de imagen, tiene un efecto relevante (d Cohen grande) respecto al promedio de satifacción y la potencia estadística es alta.

```python
## 1) cargar datos muestras: tamaño (n), media (xbarra) y sigma (S)
# muestra 1 (control) Competencia
n_1 = N_Comp
p_1 = T2BS_Comp
S_1 = des_Comp

# muestra 2 (experimental)
n_2 = N_Lepac
p_2 = T2BS_Lepac
S_2 = des_Lepac

## 2) Determinar Nivel de Significancia
alpha = 0.05

# calcular Z critico 2 colas con Scipy
Zcritico = stats.norm.ppf(1 - (alpha/2))
print (f'Z critico: {Tcritico:5.3f}')

## 3) Estimar el IC de la diferencia de proporciones
#Sigma diferencias proporciones es de la siguiente forma:
#Calcular Desv. Estandar de la diferencias de proporciones

Sp_1p_2 = np.sqrt((p_1*(1-p_1)/n_1)+(p_2*(1-p_2)/n_2))
print(f"S diferencia proporciones: {Sp_1p_2}")

IC = stats.norm.interval(1-alpha, loc=p_2 - p_1,scale=Sp_1p_2)
LI = IC[0]
LS = IC[1]
print(f"IC diferencia proporciones al {1-alpha}, NC :{IC}")

#Grafico
dif_prop = stats.norm(p_2 - p_1,Sp_1p_2)
x_dif_prop = np.linspace(dif_prop.ppf(0.01), dif_prop.ppf(0.99), 100)
pdf_dif_prop = dif_prop.pdf(x_dif_prop) #funcion de densidad probabilidad

plt.figure(figsize=(10,4))
plt.plot(x_dif_prop, pdf_dif_prop,"b",)
plt.vlines(ymin=0, ymax=14, x=LI)
plt.vlines(ymin=0, ymax=14, x=LS)
plt.legend(["Dif propor","Limit Inf","Limit Sup"], bbox_to_anchor=(1.1,1.05))
plt.show()
```

Se puede apreciar que el intervalo de Confianza de la diferencia de proporciones **incluye** el valor 0, esto quiere decir que la $H_0$ **no será rechazada**. Veamos los siguientes pasos para ratificar esto.

```python
#4) plantear hipotesis
 # H0: prop2 = prop1
 # H1: prop2 > Prop1

# 5)Calcular el estadistico Z critico a emplear
Zcritico2 = stats.norm.ppf(1-(alpha/2))
Zcritico1 = stats.norm.ppf(1-(alpha))

print(f"Z critico: {Zcritico1}")

#7) Estadistico de prueba

Z = (p_2 - p_1 -0) / Sp_1p_2
print(f"Z prueba: {Z}")

#8) Calcular p-value y comparacion con el nivel de significancia establecido.

pvalue = 1-(stats.norm.cdf(Z)) #Z prueba se encuentra en cola derecha
print(f"p-value (significancia estadistica): {pvalue}")
```

De esta manera se aprecia que la Hipotesis Nula **no es rechazada** con un 95% de confianza.

-**Analisis de potencia Le Pac / Competencia:** Aunque la Hipotesis nula no es rechazada, se realizará el análisis de potencia para medir la relevancia del efecto. 

```python
#10) Calcular la potencia de la prueba (1-beta)

#buscar el valor en Porcentaje correspondiente al Z critico
x= (Zcritico1 * Sp_1p_2)# por la derecha
print(f"X promedio para Z {Zcritico1} : {x}")

#Calcular Beta (error tipo II)
beta = stats.norm.cdf(Zcritico1-Z)
print(f"Beta : {beta}")

#Calcular Potencia estadistica (1-beta)
print(f"Potencia de la prueba : {1-beta}")

#11)Calcular el tamaño del efecto

efecto_bruto = (p_2-p_1)
print(f"diferencia bruta : {efecto_bruto}")

tamano_efecto = (efecto_bruto / sigma)
print(f'Tamaño del efecto d Cohen: {tamano_efecto}')

#Graficos
#proyectar distribucion diferencias pro H0
H0_dif_prop = stats.norm(0, Sp_1p_2)
H0_x_dif_prop = np.linspace(H0_dif_prop.ppf(0.001), H0_dif_prop.ppf(0.999), 100)
H0_pdf_dif_prop = H0_dif_prop.pdf(H0_x_dif_prop) #funcion densidad probabilidad

#proyectar distribucion diferencias pro H1
H1_dif_prop = stats.norm(p_2 - p_1, Sp_1p_2)
H1_x_dif_prop = np.linspace(H1_dif_prop.ppf(0.001), H1_dif_prop.ppf(0.999), 100)
H1_pdf_dif_prop = H1_dif_prop.pdf(H1_x_dif_prop) #funcion densidad probabilidad

plt.figure(figsize=(10,5))
plt.plot(H0_x_dif_prop, H0_pdf_dif_prop,"r",) # medias H0 fit
plt.plot(H1_x_dif_prop, H1_pdf_dif_prop,"b",) # medias H0 fit
plt.vlines(ymin=0, ymax=14, x=x)# Z critico
plt.legend(["H0","H1","Valor Z critico"], bbox_to_anchor=(1.1,1.05))
plt.show()
```
